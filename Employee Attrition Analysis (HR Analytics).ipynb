````python
import pandas as pd

# Load dataset
df = pd.read_csv("test.csv")

# Step 1: Get % of missing values in each column
missing_percent = df.isnull().mean() * 100

# Step 2: Drop columns with >75% missing
cols_to_drop = missing_percent[missing_percent > 75].index
df.drop(columns=cols_to_drop, inplace=True)
print(f"Dropped columns (>75% missing): {list(cols_to_drop)}")

# Step 3: Drop rows if column has <15% missing
cols_drop_rows = missing_percent[(missing_percent > 0) & (missing_percent < 15)].index
df.dropna(subset=cols_drop_rows, inplace=True)
print(f"Dropped rows for columns (<15% missing): {list(cols_drop_rows)}")

# Step 4: Impute columns with 20%-70% missing
cols_to_impute = missing_percent[(missing_percent >= 20) & (missing_percent <= 70)].index

for col in cols_to_impute:
    if df[col].dtype == 'object' or str(df[col].dtype) == 'category':
        # Impute with mode for categorical
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        # Impute with mean for numerical
        df[col].fillna(df[col].mean(), inplace=True)

print(f"Imputed columns (20%-70% missing): {list(cols_to_impute)}")

# Final check
print("\nRemaining missing values:\n", df.isnull().sum().sort_values(ascending=False))
````
